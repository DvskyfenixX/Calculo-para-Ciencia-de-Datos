```python
import sklearn as sk
import numpy as np
import matplotlib.pyplot as plt
from matplotlib.colors import LogNorm
from sklearn import mixture
```

```python
# Leemos nuestros datos csv
data = np.loadtxt('data_HW4.csv', delimiter=',')
data
```

$\textbf{Pregunta 1:}$ Entrene un modelo Gaussiano mixto con 2, 3 y 4 clusters con todo el set de datos. Para esto, revise el siguiente link a scikit-learn. Explíque cual de estos 3 modelos le hace más sentido viendo el scatter plot de los datos.

```python
from sklearn.mixture import GaussianMixture
# Entrenamos nuestro modelo Gaussiano mixto con 2 componentes
gmm = GaussianMixture(n_components=2)
gmm.fit(data)
predict = gmm.predict(data)
# Graficamos los datos y los centros de las componentes
plt.scatter(data[:,0], data[:,1], c=predict, cmap='plasma', edgecolors='black')
plt.scatter(gmm.means_[:,0], gmm.means_[:,1], color='red')
plt.show()
```

```python
# Entrenamos nuestro modelo Gaussiano mixto con 3 componentes
gmm = GaussianMixture(n_components=3)
gmm.fit(data)
predict = gmm.predict(data)
# Graficamos los datos y los centros de las componentes
plt.scatter(data[:,0], data[:,1], c=predict, cmap='plasma', edgecolors='black')
plt.scatter(gmm.means_[:,0], gmm.means_[:,1], color='red')
plt.show()
```

```python
# Entrenamos nuestro modelo Gaussiano mixto con 4 componentes
gmm = GaussianMixture(n_components=4)
gmm.fit(data)
predict = gmm.predict(data)
# Graficamos los datos y los centros de las componentes
plt.scatter(data[:,0], data[:,1], c=predict, cmap='plasma', edgecolors='black')
plt.scatter(gmm.means_[:,0], gmm.means_[:,1], color='red')
plt.show()
```

$\textbf{R:}$ Analizando el entrenamiento del modelo, podemos ver que a medida que aumentamos la cantidad de componentes, el error o dispersión del entrenamiento disminuye, esto debido a que al haber 4 sectores divididos o clusters de datos, al aumentar la cantidad de componentes, los centros de los componentes son más cercanos a los datos y, por lo tanto, más precisos. Caso contrario al haber una menor cantidad de componentes para dividir los datos, los centros de los componentes son más alejados y estocásticos, tendiendo a ser menos precisos para la clasificación de los datos.

$\textbf{Pregunta 2:}$ Grafique las curvas de nivel asociado a la log-verosimilitud negativa (el score con signo negativo) para cada uno de los casos anteriores. Explique por qué estos son distintos a pesar de tener los mismos datos. Use en su respuesta la información disponible sobre el modelo entrenado.

```python
# Analizamos las curvas de nivel en relación a la log-verosimilitud negativa de los datos para cada modelo

# Utilizamos el código adjuntado adjuntado en el enunciado de la tarea

# Caso 1: 2 componentes
clf = mixture.GaussianMixture(n_components=2, covariance_type="full")
clf.fit(data)
predict = clf.predict(data)
# display predicted scores by the model as a contour plot
x = np.linspace(data[:, 0].min(), data[:, 0].max())
y = np.linspace(data[:, 1].min(), data[:, 1].max())
X, Y = np.meshgrid(x, y)
XX = np.column_stack([X.ravel(), Y.ravel()])
Z = -clf.score_samples(XX)
Z = Z.reshape(X.shape)

CS = plt.contour(X, Y, Z, levels=20, cmap='viridis')
CB = plt.colorbar(CS, shrink=0.8, extend="both")
plt.scatter(data[:, 0], data[:, 1], c=predict, cmap='plasma')
plt.scatter(clf.means_[:, 0], clf.means_[:, 1], color='red')

plt.title("Curvas de nivel asociadas a la log-verosimilitud negativa de los datos")
plt.show()
```
